```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
forestfire=pd.read_csv("forestfires.csv")
forestfire
```

```python
forest=forestfire.drop(forestfire.columns[[0,1]],axis=1)
forest
```

```python
y1=forestfire.iloc[:,-1:]
y1
```

```python
y2=pd.get_dummies(y1)
y2
```

```python
y=y2.iloc[:,0:1]
y
```

```python
x=forest.iloc[:,:-1]
x
```

```python
import seaborn as sns
sns.pairplot(x,hue=None)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_x=sc.fit_transform(train_x)
test_x=sc.transform(test_x)
```

```python
#creating a model for SVM
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=1)
classifier.fit(train_x,train_y)
```

```python
y_pred=classifier.predict(test_x)
y_pred
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,y_pred)
```

```python
classifier.score(test_x,test_y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y,y_pred)
```

```python
#creating a model for SVM
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=1)
classifier.fit(train_x,train_y)
```

```python
y_pred1=classifier.predict(test_x)
y_pred1
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,y_pred1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y,y_pred1)
```
