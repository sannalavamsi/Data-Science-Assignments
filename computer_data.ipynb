```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
computers=pd.read_csv("Computer_Data.csv")
computers.head()
```

```python
A=computers.iloc[:,1:]
A
```

```python
A.shape
```

```python
A.isnull().sum()
```

```python
A.corr()['price']
```

```python
Y=A.iloc[:,0:1]
Y
```

```python
X=A.iloc[:,1:]
X
```

```python
S=pd.get_dummies(X)
S
```

```python
C=X.drop(X.columns[[4,5,6]],axis=1)
C
```

```python
from sklearn.model_selection import train_test_split
S_train,S_test,Y_train,Y_test=train_test_split(S,Y,test_size=0.2,random_state=0)
S_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(S_train,Y_train)
```

```python
Y_pred=regression.predict(S_test)
Y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_pred,Y_test)
score
```

```python
from sklearn.model_selection import train_test_split
S_train,S_test,Y_train,Y_test=train_test_split(np.sqrt(S),Y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(S_train,Y_train)
Y_pred=regression.predict(S_test)
Y_pred
from sklearn.metrics import r2_score
score=r2_score(Y_pred,Y_test)
score
```
