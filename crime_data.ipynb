```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime=pd.read_csv("crime_data.csv")
crime
```

```python
#Normalization
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df=norm_func(crime.iloc[:,1:])
df
```

```python
df.corr()
```

```python
#Dendrogram
import scipy
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#  dendrogram linkage="ward"
z=linkage(df,method='ward',metric='euclidean')
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    z,
    leaf_rotation=90,  # rotates the x axis labels
    leaf_font_size=8  # font size for the x axis labels
)
plt.show()
```

```python
#  dendrogram linkage="complete"
z=linkage(df,method='complete',metric='euclidean')
plt.figure(figsize=(25,10))
plt.title('hierarchical clustring')
plt.xlabel('index')
plt.ylabel('distance')
dendrogram(z,leaf_rotation=90,leaf_font_size=8)
plt.show()
```

```python
#  dendrogram linkage="average"
z=linkage(df,method='average',metric='euclidean')
plt.figure(figsize=(25,10))
plt.title('hierarchical clustring')
plt.xlabel('index')
plt.ylabel('distance')
dendrogram(z,leaf_rotation=90,leaf_font_size=8)
plt.show()
```

```python
# AgglomerativeClustering ,linkage="ward"
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=3,linkage='ward',affinity='euclidean')
y_hc=clustering.fit_predict(df)
y_hc
```

```python
#scatter plot
X=np.array(df)
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='blue')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='yellow')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='red')
plt.show()
```

```python
# AgglomerativeClustering ,linkage="complete",Scatterplot
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity='euclidean')
y_hc=clustering.fit_predict(df)
X=np.array(df)
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='blue')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='yellow')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='red')
plt.show()
```

```python
# AgglomerativeClustering ,linkage="average",scatterplot
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=4,linkage='average',affinity='euclidean')
y_hc=clustering.fit_predict(df)
X=np.array(df)
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='blue')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='yellow')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='red')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='green')
plt.show()
```
