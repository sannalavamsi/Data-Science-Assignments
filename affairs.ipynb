```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("affairs.csv")
df
```

```python
df.isnull().sum()
```

```python
df['affairs'].unique()
```

```python
def affairs(dataframe):
    dataframe.loc[df['affairs']%2==0,'affairs']=0
    dataframe.loc[df['affairs']%2!=0,'affairs']=1
    
    return dataframe
affairs(df)
```

```python
df.head()
```

```python
y=df.iloc[:,0:1]
y
```

```python
x=df.iloc[:,1:9]
x
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x=LabelEncoder()
x['gender']=labelencoder_x.fit_transform(x['gender'])
x['children']=labelencoder_x.fit_transform(x['children'])
x
```

```python
A=pd.concat([y,x],axis=1)
A
```

```python
A.corr()['affairs']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logisticregression=LogisticRegression()
logisticregression.fit(train_x,train_y)
```

```python
y_pred=logisticregression.predict(test_x)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,test_y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,test_y)
```
